```python
from SpringTransportationAPI import stAPI
from st_chart import stChart
from pyecharts.charts import Tab
import pandas as pd
```

```python
file_path1 = "D:/Resource/春运火车票余票信息查询数据.csv"
file_path2 = "D:/Resource/cnstation.csv"
st = stAPI(pd.read_csv(file_path1, encoding='gbk'), pd.read_csv(file_path2, encoding='gbk'))
```

```python
# 分析车次覆盖情况（任务8）
# 所有的一线城市
cities = set(st.st_df['dep_city'].tolist())
# 声明stChart对象
charts = stChart()
# 声明pyecharts的Tab对象
tab = Tab()

# 通过循环来加入19个一线城市的数据
for c in cities:
    tmp_c = st.train_track(c)
    tmp_p = charts.train_track(tmp_c, c)

    tab.add(tmp_p, c)

tab.render_notebook()
```

```python
# 分析全国站点分布（任务7）
r_t_c_dep, r_t_c_des = st.train_count()
data_collection = [st.convert_data(r_t_c_dep), st.convert_data(r_t_c_des)]

charts = stChart()
p = charts.port_dis(data_collection)
p.render_notebook()
```

```python
# 分时间段的车次情况
d_t = ['0-6', '6-12', '12-18', '18-24']
day = [114, 115, 117, 118]

x = []
for i in day:
    tmp = []
    for j in d_t:
        b, c = st.limit_time(j, i)
        tmp.append(b)

    x.append(tmp)

charts = stChart()
p = charts.bar_chart(x)
p.render_notebook()
```

```python
# 统计春运列车情况（任务6）
a = st.stock_analysis()
name = []
value = []

for i in a:
    name.append(i[0])
    value.append(round(i[1], 2))

charts = stChart()
p = charts.bar_chart_stock(name, value)
p.render_notebook()
```
